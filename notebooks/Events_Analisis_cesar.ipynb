# ¿Cual es la relacion entre los dispositivos y eventos?

```python
dfEvents.groupby(['application_id','event_id'])['total'].count().to_frame().sort_values(by=['application_id'],ascending=False)
```

```python
dfEvents_gropuby = dfEvents.groupby(['application_id','event_id'])['total'].count()
dfEvents_gropuby
```

```python
dfPivot = dfEvents_gropuby.reset_index().pivot(index='event_id', columns='application_id', values='total')
dfPivot = dfPivot.fillna(0)
```

```python
dfPivot.head(10)
```

```python
colors = ["#006D2C", "#31A354","#74C476"]
dfPivot.plot.bar(stacked=True, figsize=(12,10))
plt.title('Las modelos de celuares por marca que mas apps instalaron', fontsize=20);
plt.ylabel('Cantidad de instalaciones', fontsize=16);
plt.xlabel('Marca', fontsize=20);
plt.show()
```

```python
fig, ax = plt.subplots(figsize=(14,16))         # Sample figsize in inches
sns.heatmap(dfPivot, annot=False, linewidths=.5010, ax=ax); #linewidths=.00010,
```
